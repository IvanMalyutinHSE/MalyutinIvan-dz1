```python
#при alpha = 0.1
import sympy
x = sympy.Symbol('x')
N = 25
exact = [float(sympy.integrate(x**n / (x + 0.1), (x, 0, 1))) for n in range(N)]
print(exact)
```

```python
#вычисляем прямой рекурсией I_25(alpha = 10)
import numpy as np

alpha = 10
I = np.log((1 + alpha)/alpha)
i = 0
while i <= 24:
    i += 1
    I = (1 / i) - (alpha * I)
print(I)
#как мы видим, результат при больших alpha значительно отличается от реального ответа(получилось большое отрицательное число)
```

```python
#вычисляем прямой рекурсией I_25(alpha = 0.1)
import numpy as np

alpha = 0.1
I = np.log((1 + alpha)/alpha)
i = 0
while i <= 24:
    i += 1
    I = (1 / i) - (alpha * I)
print(I)
#получился ответ, близкий к реальному (небольшое положительное значение)
#можно сделать вывод о том, что при малых alpha(со значениями меньше 1 алгоритм работает корректно, в отличие от больших значений alpha(в примере alpha =10))
```

```python
#вычисляем обратной рекурсией I_25(alpha = 10)
import numpy as np

alpha = 10
I = 0
i = 100
while i >= 26:
    i -= 1
    I = ((1 / ((alpha)*(i+1))) - (I/alpha))
print(I)
#мы получили небольшое положительное число, что означает, что обратная рекурсия для alpha > 1 работает правильно
```
